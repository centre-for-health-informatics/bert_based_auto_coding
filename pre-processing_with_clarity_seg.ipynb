```python
import os
import imp
import sys
import re
sys.path.append('path_to_ClarityNLP/ClarityNLP/nlp')
sys.path.append('path_to_ClarityNLP/ClarityNLP/')

import Sentence_tokenizer_from_Clarity as cla_token
import time
from tqdm import tqdm

def back_to_doc_seg_with_clarity(ratio = 2):
    src_dir = './pu_data_ratio' + str(ratio) +'/'
    text_train = pickle.load(open(src_dir+'text_train.pkl','rb'))
    label_train = pickle.load(open(src_dir+'label_train.pkl','rb'))
    text_test = pickle.load(open(src_dir+'text_test.pkl','rb'))
    label_test = pickle.load(open(src_dir+'label_test.pkl','rb'))
    des_dir = './pu_data_ratio_Clarity' + str(ratio) +'/'
    if not os.path.exists(des_dir):
        os.mkdir(des_dir)
    doc_text_train = back_to_doc(text_train)
    doc_text_test = back_to_doc(text_test)
    new_text_train = []
    new_text_test = []
    for text in tqdm(doc_text_train):
        text = re.sub(r'\s*\[.*?\]\s*',' ',text)
        list_of_sent = cla_token.parse_sentences_spacy(text)
        new_text_train.append(list_of_sent)
    pickle.dump(new_text_train,open(des_dir+'text_train.pkl','wb'))
    
    for text in tqdm(doc_text_test):
        text = re.sub(r'\s*\[.*?\]\s*',' ',text)
        list_of_sent = cla_token.parse_sentences_spacy(text)
        new_text_test.append(list_of_sent)
    pickle.dump(new_text_test,open(des_dir+'text_test.pkl','wb'))
    
    print('Done')
    return new_text_train,new_text_test
```

```python
import pickle
from sklearn.model_selection import train_test_split

def re_split_by_startify(ratio = 2):
    src_dir = './pu_data_ratio_Clarity' + str(ratio) +'/'
    text_train = pickle.load(open(src_dir+'text_train.pkl','rb'))
    label_train = pickle.load(open(src_dir+'label_train.pkl','rb'))
    text_test = pickle.load(open(src_dir+'text_test.pkl','rb'))
    label_test = pickle.load(open(src_dir+'label_test.pkl','rb'))
    text_total = text_train
    text_total.extend(text_test)
    label_total = label_train
    label_total.extend(label_test)
    des_dir = src_dir
    X_train, X_test, y_train, y_test = train_test_split(text_total, label_total, test_size=0.2, random_state=42,stratify=label_total)
    pickle.dump(X_train,open(des_dir+'text_train_stratified.pkl','wb'))
    pickle.dump(X_test,open(des_dir+'text_test_stratified.pkl','wb'))
    pickle.dump(y_train,open(des_dir+'label_train_stratified.pkl','wb'))
    pickle.dump(y_test,open(des_dir+'label_test_stratified.pkl','wb'))
    print('Done')
```
